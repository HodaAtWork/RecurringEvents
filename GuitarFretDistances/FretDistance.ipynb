# Guitar fret distance calculation 
Playing classical guitar I hardly pay attention to the position of the frets. However it is interesting to know why frets are located on specific fretboard positions. 

Frets are the raised, metals strips placed on the neck of the guitar. The first fret is the strip closest to the guitar's headstock. Frets are fixed on the neck's fretboard and divide the neck into fixed segments at intervals related to a musical framework. Each fret represents one semitone in the standard western system, in which one octave is divided into twelve semitones. Moving from the nut each fret raises the note by such a semitone until a full octave is reached at the 12th fret. It is this 12th fret that divides the distance between nut (or sometimes 0-fret) and bridge, the scale length, exactly in half. <br><br> Headstock and fretboard with 0-fret:<br>

### 12-TET
Dividing the octave into 12 parts with a ratio of distances between two frets of $\sqrt[12]{2}$ is called 12-tone equal temperament, abbreviated 12-TET. On a logarithmic scale all 12 parts are equal.<br>The exact calculation of equal temperament is attributed to Zhu Zaiyu (China, 1584) and Simon Stevin (Flanders, 1585).

### Calculator fret positions
Calulating the fret positions is quite easy. Presented are two ways to do so. 
* Using the equal temperament constant of 17.817. This constant is derived from the ratio as follows: $\frac{scalelength}{scalelength  -  (\frac{scalelength}{\sqrt[12]{2}})}$
* Using the ratio $\sqrt[12]{2}$ between two frets.

```python
scale_length = 650
#constant = 17.817
constant = scale_length / (scale_length - (scale_length / 2**(1/12)))
from_nut = 0
for fret in range(1,13):
    # constant
    from_nut = (scale_length - from_nut)/constant + from_nut
    # ratio
    d = scale_length - (scale_length / (2**(fret/12)))
    print('fret ', fret, ' distance from nut Method 1: ', from_nut, ' Method 2: ', d)
```

```python
fig, ax = plt.subplots(figsize=(12,10))
parts12Tet=[]
for tone in range(1,13):
    parts12Tet.append(2**(tone/12))
labels = ['minor 2nd', 'major 2nd', 'minor third', 'major 3rd', 'perfect 4th', 'tritone', 'perfect 5th', 'minor 6th', 'major 6th', 'minor 7th', 'major 7th', 'octave']
bar_plot=plt.bar(range(1,13), parts12Tet, width=.3, color='b')
for i, rect in enumerate(bar_plot):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
            labels[i],
            ha='center', va='bottom', rotation=0)
plt.title('Twelve-tone equal temperament', fontsize=15)
plt.xlabel('Octave', fontsize=15)
plt.yscale('log')
plt.ylabel('Value in 12-Tet (log)', fontsize=15)
plt.show()
print(parts12Tet)
```
